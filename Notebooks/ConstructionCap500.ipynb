```python
cap500_names = cap500.set_index("species_name")
display(cap500_names)
cap500_cropped_names = cap500_cropped.set_index("species_name")
display(cap500_cropped_names)
```

```python
#dic = {"species_name" : [], "species_num" : [], "predicted_species_name" : [], "predicted_species_num" : [], "good_pred_probability" : [], "img_name" : [], "img_path" : []}
dic = {"species_name" : [], "path_to_img" : []}
data_val = pd.DataFrame(dic)
data_test = pd.DataFrame(dic)
data_train = pd.DataFrame(dic)
dic_cap = {"species_name" : [], "path_to_img" : []}
data_val_cap = pd.DataFrame(dic_cap)
data_test_cap = pd.DataFrame(dic_cap)
data_train_cap = pd.DataFrame(dic_cap)
# display(data_test)
# display(especes)
# display(cap500_names)
```

```python
'''idée: prendre au hasard les images pour générer ce premier train-data-set. Puis pour ceux des images capées on fait en fonction de ces résultats'''

#Création dy train-valid-test du non cappé
for i in range(nb_especes):
    species_name = especes.at[i, "species"]
    df = cap500_names.loc[species_name]
    df.reset_index(inplace=True)
   
    #display(df)
    nb_images = len(df)
    nb_images_test = int(0.1 * nb_images) #on prend 10 pourcent du nombre d'images du cap 500
    # Ensemble de test
    sample_test = df.sample(nb_images_test) #on choisit au hasard ces 10% d'images
    # Suppression des lignes pour éviter d'avoir des images à la fois dans le train et dans la validation:ok grace à merge et 'left_only'
    df = df.merge(sample_test, how='left', indicator=True) #concatène deux dataframes pour les comparer
    #display(df)
    df = df[df['_merge'] == 'left_only'] #on garde les lignes qui ne sont pas dans le sample_test car sinon il y a marqué 'both'
    df.drop(columns="_merge", inplace=True) #on supprime la colonne _merge
    df.reset_index(inplace=True)
    df.drop(columns="index", inplace=True)
    
    # Ensemble de validation
    sample_val = df.sample(nb_images_test)
    # Suppression des lignes pour la même raison que ci-dessus
    df = df.merge(sample_val, how='left', indicator=True)
    df = df[df['_merge'] == 'left_only']
    df.drop(columns="_merge", inplace=True)
    df.reset_index(inplace=True)
    df.drop(columns="index", inplace=True)
    
    
    data_test = pd.concat([data_test, sample_test])#on remplit progressivement le data_test espèce par espèce
    data_val = pd.concat([data_val, sample_val])
    data_train = pd.concat([data_train, df])

display(data_test)  
display(data_val)
display(data_train)

data_test.reset_index(inplace=True)
data_test.drop(columns="index", inplace=True)
data_val.reset_index(inplace=True)
data_val.drop(columns="index", inplace=True)
data_train.reset_index(inplace=True)
data_train.drop(columns="index", inplace=True)
```

```python
#On crée un train-valid-set pour les images croppées avec les mêmes images que pour le non cappé

good_image_name_test = data_test["image_name"]
good_image_name_val = data_val["image_name"]
good_image_name_train = data_train["image_name"]

for i in range(nb_especes):
    species_name = especes.at[i, "species"]
    df_cap = cap500_cropped_names.loc[species_name]
    df_cap.reset_index(inplace=True)
    nb_images = len(df_cap)
    nb_images_test = int(0.1 * nb_images) #on prend 10 pourcent du nombre d'images du cap 500

    # Ensemble de test
    #sample_test = df_cap.sample(nb_images_test) #on choisit au hasard ces 10% d'images
    sample_test = df_cap[df_cap["image_name"].isin(good_image_name_test)] #on ne prend que les images qui sont déjà dans le test
    
    # Suppression des lignes pour éviter d'avoir des images à la fois dans le train et dans la validation:ok grace à merge et 'left_only'
    df_cap = df_cap.merge(sample_test, how='left', indicator=True) #concatène deux dataframes pour les comparer
    df_cap = df_cap[df_cap['_merge'] == 'left_only'] #on garde les lignes qui ne sont pas dans le sample_test car sinon il y a marqué 'both'
    df_cap.drop(columns="_merge", inplace=True) #on supprime la colonne _merge
    df_cap.reset_index(inplace=True)
    df_cap.drop(columns="index", inplace=True)
    
    # Ensemble de validation
    #sample_val = df_cap.sample(nb_images_test)
    sample_val = df_cap[df_cap["image_name"].isin(good_image_name_val)]#on ne prend que les images qui sont déjà dans le test
    # Suppression des lignes pour la même raison que ci-dessus
    df_cap = df_cap.merge(sample_val, how='left', indicator=True)
    df_cap = df_cap[df_cap['_merge'] == 'left_only']
    df_cap.drop(columns="_merge", inplace=True)
    df_cap.reset_index(inplace=True)
    df_cap.drop(columns="index", inplace=True)
    
    
    data_test_cap = pd.concat([data_test_cap, sample_test])#on remplit progressivement le data_test espèce par espèce
    data_val_cap = pd.concat([data_val_cap, sample_val])
    data_train_cap = pd.concat([data_train_cap, df_cap])

display(data_test_cap)  
display(data_val_cap)
display(data_train_cap)

data_test_cap.reset_index(inplace=True)
data_test_cap.drop(columns="index", inplace=True)
data_val_cap.reset_index(inplace=True)
data_val_cap.drop(columns="index", inplace=True)
data_train_cap.reset_index(inplace=True)
data_train_cap.drop(columns="index", inplace=True)


```

```python
#on créé des répertoires vides pour enregistrer les images

path = "/home/lucien/Documents/projet_abeilles/abeilles-cap500"
name_directory = "cap500"

def create_species_directory(path, name_directory, especes_df):
    
    nb_especes = len(especes_df)
    
    directory_path = path + "/" + name_directory
    os.mkdir(directory_path)
    os.mkdir(directory_path + "/test")
    os.mkdir(directory_path + "/train")
    os.mkdir(directory_path + "/val")
    
    for i in range(nb_especes):
        species_name = especes_df.at[i, "species"]
        os.mkdir(directory_path + "/test/" + species_name)
        os.mkdir(directory_path + "/train/" + species_name)
        os.mkdir(directory_path + "/val/" + species_name)

def create_species_directory_cap(path, name_directory, especes_df):
    
    nb_especes = len(especes_df)
    
    directory_path = path + "/" + name_directory
    os.mkdir(directory_path + "/test_cap")
    os.mkdir(directory_path + "/train_cap")
    os.mkdir(directory_path + "/val_cap")
    
    for i in range(nb_especes):
        species_name = especes_df.at[i, "species"]
        os.mkdir(directory_path + "/test_cap/" + species_name)
        os.mkdir(directory_path + "/train_cap/" + species_name)
        os.mkdir(directory_path + "/val_cap/" + species_name)
```

```python
from PIL import Image
def save_images(data, path_bd, path_destination, nature):
    
    nb_images = len(data)
    

    for i in range(nb_images):
        relative_img_path = data.at[i, "path_to_img"]
        img_name = data.at[i, "image_name"]
        species = data.at[i, "species_name"]
        imgpil = Image.open(relative_img_path)
        imgpil2 = imgpil.resize((224,224))
        #img = np.array(imgpil)
        #imgpil2 = Image.fromarray(img)
        name_to_save = path_destination + "/" + nature + "/" + species + "/" + img_name
        if not os.path.isfile(name_to_save):
            imgpil2.save(name_to_save)
        else:
            print("Problème, image déjà présente...")
```

```python
path_bd = "/home/acarlier/Desktop/CAP600"

#pour les images non croppéees
save_images(data_val, path, path + "/cap500", "val")
save_images(data_test, path, path + "/cap500", "test")
save_images(data_train, path, path + "/cap500", "train")

#pour le simages croppées
save_images(data_val_cap, path, path + "/cap500", "val_cap")
save_images(data_test_cap, path, path + "/cap500", "test_cap")
save_images(data_train_cap, path, path + "/cap500", "train_cap")
```
