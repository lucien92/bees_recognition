# Corrélation F1-score / support ?

```python
path = "/content/drive/MyDrive/Stage2A"
```

## Chargement du csv

```python
import pandas as pd

data = pd.read_csv(path + "/cap500/ResNet101/ResNet101_Sequence/evaluation_by_species_resnet101_sequence_cap500.csv")
display(data)
```

```python
f1_score = []
support = []
for i in range(71):
  f1_score.append(data.at[i, "f1-score"])
  support.append(data.at[i, "support"])
```

```python
import matplotlib.pyplot as plt
plt.scatter(support, f1_score, c='red', marker = '+')
plt.xlabel("Support")
plt.ylabel("F1_score")
plt.axhline(y=data.at[72, "f1-score"], color='b', linestyle='-', label="macro avg")
plt.legend(loc="lower right")
plt.show()
#plt.savefig(path + "/cap500/ResNet101/ResNet101_Sequence/figure1.jpg")
```
